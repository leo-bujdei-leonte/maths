```python
import numpy as np
from scipy.sparse import coo_matrix
import scipy as sp
from scipy.sparse.linalg import svds

d = np.array([3, 4, 5, 7, 2, 6]) 
r = np.array([0, 0, 1, 1, 3, 3]) 
c = np.array([2, 4, 2, 3, 1, 2]) 
b = []
coo = coo_matrix((d, (r, c)), shape=(4, 5))

u, s, vt = svds(coo, k=min(coo.shape)-1, which='SM')

tol = 1e-12
null_mask = s < tol
null_vec = vt.T[:, null_mask]

prod = coo @ null_vec

print("coo", coo, "\nvt.t", vt.T, "\ns", s, "\nnull_vec", null_vec, sep="\n")
print(np.allclose(prod, np.zeros(prod.shape), tol))
# print(coo.toarray())
```
